## Tokenization

Tokenization breaks down a text sequence into fragments.
NLTK has its own tokenization methods.

````python
from nltk.tokenize import word_tokenize
sent = '''I amd reading a book. 
It is Python Machine Learning by Example, 
2nd edition.'''
print(word_tokenize(sent))
````

````python
sent2 = "I have been to U.K. and U.S.A."
print(word_tokenize(sent2))
````

We can also use spaCy to tokenize. SpaCy's tokenization function identifies acronyms.

````python
import spacy
nlp = spacy.load('en_core_web_sm')
tokens2 = nlp(sent2)
print([token.text for token in tokens2])
````

We can also tokenize text by sentence.

````python
from nltk.tokenize import sent_tokenize
print(sent_tokenize(sent))
````

## POS Tagging

Part-of-speech tagging can be used to get a grammatical category of each word (token) in a sentence (or document).

````python
import nltk
from nltk.tokenize import word_tokenize
sent = '''I amd reading a book. 
It is Python Machine Learning by Example, 
2nd edition.'''
tokens = word_tokenize(sent)
print(nltk.pos_tag(tokens))
````

NLTK provides help functions to identify what each POS tag means.

````python
nltk.help.upenn_tagset('PRP')
nltk.help.upenn_tagset('VBP')
nltk.help.upenn_tagset('NNP')
nltk.help.upenn_tagset('NN')
nltk.help.upenn_tagset('CD')
nltk.help.upenn_tagset('DT')
````

SpaCy also includes POS tagging, inside the pos_ elment of the nlp() method.

````python
import spacy
nlp = spacy.load('en_core_web_sm')
sent2 = "I have been to U.K. and U.S.A."
tokens2 = nlp(sent2)
print([(token.text,token.pos_) for token in tokens2])
````

## Named-entity recognition

Given a text sequence, the named-entity recognition (NER) task is to locate and identify words or phrases that are of definitive categories such as names of persons, companies, locations, and dates.

````python
import spacy
nlp = spacy.load('en_core_web_sm')
tokens3 = nlp('The book written by Hayden Lui in 2018 was sold at $30 in America.')
print([(token_ent.text,token_ent.label_)
       for token_ent in tokens3.ents])
````

## Feature extraction

### Counting the occurence of each word token
We can count the most frequent words in the data using the CountVectorizer function from sci.kit learn.

````python
from sklearn.feature_extraction.text import CountVectorizer
````

We'll limit to the 500 most frequent words.

````python
count_vector = CountVectorizer(max_features=500)
data_count = count_vector.fit_transform(groups.data)
data_count
````

This created a token count matrix to identify which documents contain the top words and how frequent each word is in each document.
<br>
Let's see the count of each top word for the first document:

````python
data_count.toarray()[0]
````

Now let's see which are the 500 top words.

````python
print(count_vector.get_feature_names())
````

### Text preprocessing
This list of 500 word sis not really helpful since it contians numbers.
<br>
Let's start over but we'll clean the data first.
<br>
We'll define a function to recognize letter-only words.

````python
def is_letter_only(word):
    for char in word:
        if not char.isalpha():
            return False
    return True
````

Now let's generate a cleaned version of the newsgroup data.

````python
data_cleaned = []
for doc in groups.data:
    doc_cleaned = ' '.join(word for word in doc.split()
                          if is_letter_only(word))
    data_cleaned.append((doc_cleaned))
````

### Dropping stop words
Now we'll use stop words to drop words that might be common but not provide value.

````python
from sklearn.feature_extraction import stop_words
print(stop_words.ENGLISH_STOP_WORDS)
````

Use the stop_word parameter in CountVectorizer to use the stop words.

````python
count_vector_sw = CountVectorizer(stop_words="english",max_features=500)
````

### Stemming and Lemmatizing
Now we'll run the complete process:
- Preprocessing
    * Removing names
    * Removing stop-words
- Lemmatizing
- Count vectorizing

````python
from nltk.corpus import names
all_names = set(names.words())
count_vector_sw = CountVectorizer(stop_words='english',max_features=500)
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
data_cleaned = []
for doc in groups.data:
    doc = doc.lower()
    doc_cleaned = ' '.join(lemmatizer.lemmatize(word) 
                           for word in doc.split()
                           if is_letter_only(word) and
                           word not in all_names)
    data_cleaned.append(doc_cleaned)
data_cleaned_count = count_vector_sw.fit_transform(data_cleaned)
````

Now, the features (top 500 words) seem more meaningful.

````python
print(count_vector_sw.get_feature_names())
````

## Visualizing the newsgroups data with t-SNE
We now have 500 features. But maybe some have more value than others when trying to use them to distinguish between newsgroups.
<br>
We can use t-SNE for dimensionality reduction.

````python
from sklearn.manifold import TSNE
````

On this example, we'll only work with three categories.

````python
categories_3 = ['talk.religion.misc','comp.graphics','sci.space']
groups_3 = fetch_20newsgroups(categories=categories_3)
````

Then we apply the t-SNE to reduce the 500-dimensional matrix to a two-dimensional matrix.
The parameters we specify in the TSNE object are as follows:
- *n_components*: The output dimension
- *perplexity*: The number of nearest data points considered neighbors in the algorithm with a typical value of between 5 and 50
- *random_state*: The random seed for program reproducibility
- *learning_rate*: The factor affecting the process of finding the optimal mapping space with a typical value of between 10 and 1,000

````python
tsne_model = TSNE(n_components = 2, perplexity = 40,
                  random_state=42,learning_rate=500)
data_tsne = tsne_model.fit_transform(data_cleaned_count.toarray())
````

````python
import matplotlib.pyplot as plt
plt.scatter(data_tsne[:,0],data_tsne[:,1],c=groups_3.target)
plt.show()
````
